Ejercicio: Considere $p_m = 2p_{m-1} - p_{m-2}$, esta ecuacion recursiva tiene solucion general $p_m = c_1 + c_2m$. Considere las c. i., $p_0 = 1$ y $p_1 = 1/3$, determine $p_m$ y $\hat{p}_m$ para este caso particular y para un redondeo a cuatro cifras. Muestre que el error es lineal.

Sea $$p_m = c_1+c_2m$$
$$p_2 = 2\frac{1}{3}-1$$
$$= c_1 + c_2(2)$$
$$\hat{c}_1 = 1$$
$$\hat{c}_2 = 0.6666$$
$$p_m = 2p_m + 2p_{m-1} - p_{m-2}$$
$$0 = -p_m + 2p_{m-1}-p_{m-2}$$
$$0 = p^2 - 2p + 1$$
$$=(p-1)^2$$
$$\therefore p = 1$$
$$p_m = 1 - 0.6666\cdot m$$
$$\therefore p_m = 0.3334$$

```python
import numpy as np 
import matplotlib.pyplot as plt 

m = np.linspace(0, 15, 200)
error = 0.3334e-4*m 

plt.figure()
plt.plot(m, error)
plt.xlabel("m")
plt.ylabel(r"$p_m-\hat{p}_m$")
plt.title("Grafica de error por redondeo")
plt.grid()
plt.show()
```

Por lo tanto el error es lineal
